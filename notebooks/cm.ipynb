```python
import pandas as pd

df = pd.read_csv("expectativa_de_vida.csv")
display(df)
```

```python
lsq = metnum.LeastSquareMethod()
```

```python
import numpy as np

unos = np.ones(183)
x = df["Adult Mortality"].values.reshape(-1, 1)
y = df["Life expectancy"].values.reshape(-1, 1)
matriz = np.column_stack((unos, x))
```

```python
lsq.ajustar(matriz, y)
```

```python
matriz.shape
```

```python
import statsmodels.api as sm
```

```python
regresion = sm.OLS(y, matriz).fit()
```

```python
regresion.params
```
