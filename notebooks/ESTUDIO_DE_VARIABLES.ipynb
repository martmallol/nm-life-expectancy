# Análisis Exploratorio de Datos

## Cargo los datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("expectativa_de_vida.csv")
display(df)
```

# Estudio de variables

### Graficos generales

```python
df = df.rename(columns={'HIV/AIDS': "HIV-AIDS"})
columnas_f = df.columns
y_labels = {
    'Country':'country',
    'Unnamed: 0': '0',
    'Status':'desarrollado o en desarrollo',
    'Life expectancy' : 'Esperanza de vida en años',
    'Adult Mortality' : 'Cantidad de muertos de entre 15 y 60 años cada 1000 habitantes',
    'infant deaths' : 'Cantidad de infantes muertos cada 1000 habitantes',
    'Alcohol' : 'Cantidad de consumo de alcohol en litros per capita',
    'percentage expenditure' :' porcentage del pbi empleado en salud',
    'Hepatitis B' : 'inmunización de Hepatitis B en menores a 1 año cada 1000 habitantes',
    'Measles' : 'Cantidad de casos de sarampion cada 1000 habitantes',
    'BMI': 'Promedio de la base corporal de la población',
    'under-five deaths' : 'Numero de muertes de menores de 5  años cada 1000 habitantes',
    'Polio' : 'Porcentage de inmunización en menores de 1 año de Polio',
    'Total expenditure' : 'Porcentage empleado en salud sobre el porcentage de GE',
    'Diphtheria' : 'inmunización entre bebes menores a 1 año de Diphtheria',
    'HIV-AIDS' : 'Cantidad de muertes DE HIV/AIDS cada 1000 nacimientos',
    'GDP' : 'PBI per capita (en USD)',
    'Population' : 'Cantidad de población',
    'thinness  1-19 years': 'Prevalencia de delgadez en población entre 10 y 19 años',
    'thinness 5-9 years' : 'Prevalencia de delgadez en población entre 5 y 9 años',
    'Income composition of resources' : 'Que tan productivamente se utilizan los recursos',
    'Schooling' : 'Cantidad de años en la escuela'
}

for c in columnas_f:
    fig = sns.displot(df, x=c, bins=40)
    plt.suptitle(c, fontsize=15)
    plt.ylabel("Cantidad de paises")
    plt.xlabel(y_labels[c])
    fig.figure.savefig(str(c) + ".png")

    
```
